# Hangman Game

```python
import random 

keep_playing = "yes"

while keep_playing == "yes":
    word_list = ["fruit", "artificial", "Bus", "bus", "pakistan"]
    chosen_word = random.choice(word_list).lower()  

    guessed_letters = []   
    chances  = 6      

    hangman_stages = [
        """
-------
|     |
      |
      |
      |
      |
-------
""",
        """ 
-------
|     |
o     |
      |
      |
      |
-------
""",
        """
-------
|     |
o     |
|     |
      |
      |
-------
""",
        """ 
------- 
|     |
o     |
/|    |
      |
      |
-------
""",
        """ 
------- 
|     |
o     |
/|\\   |
      |
      |
-------
""",
        """ 
------- 
|     |
o     |
/|\\   |
/     |
      |
-------
""",
        """
------- 
|     |
o     |
/|\\   |
/ \\   |
      |
-------
"""
    ]

    print("Hey there! Welcome to the Hangman Game. let's see if you can find the word. GOOD LUCK!")

    while chances > 0:
        current_display = ""
        for ch in chosen_word:
            if ch in guessed_letters:
                current_display += ch   
            else:
                current_display += "- " 

        print("\nWord:", current_display)

        if "- " not in current_display:
            print(" Great! You actually guessed it. The word was:", chosen_word)
            break

        user_guess = input("Enter a letter: ").lower()
        print(" You typed:", user_guess)

        if user_guess in guessed_letters:
            print(" You already tried that one...")
        elif user_guess in chosen_word:
            guessed_letters.append(user_guess)
            print("Nice! That letter is in the word.")
        else:
            chances -= 1
            print("No that is not correct. You have got", chances, "tries left.")
            print(hangman_stages[6 - chances])

    if chances == 0:
        print("Game over... You lost. The word was:", chosen_word)

    keep_playing = input("\nPlay again? (yes/no): ").lower()

print("Alrighty, thanks for playing! See you next time ")
```
